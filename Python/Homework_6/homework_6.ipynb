```python
import numpy as np
import pandas as pd

# Загрузите набор данных фильмов в pandas
database_films = pd.read_csv('https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv')
database_films
```

```python
# информация об индексах 

database_films.columns
```

```python
# запустив данную команду проверили сколько всего строк и сколько из них имеют значение NaN (таких нет)
# а также типы данных для каждого столбца и объем занимаемой памяти
database_films.info()
```

```python
# Worldwide Gross имеет тип строка, но там есть данные цифровые по которым тоже интересует аналитика
# делаем преобразование типа str в float для данного столбцa
# вариант 1
database_films['Worldwide Gross'] = database_films['Worldwide Gross'].apply(lambda x: x.replace('$', '')
                              if isinstance(x, str) else x ).astype(float)

# вариант 2
# database_films.loc[:, ('Worldwide Gross')] = database_films['Worldwide Gross'].str.replace('$','',regex=False)	
# database_films['Worldwide Gross'] = database_films['Worldwide Gross'].astype(np.float64)

database_films
```

```python
# проверили на 100% дубликаты
database_films.value_counts()

```

```python
# показало что есть дубликаты по названию
database_films.describe(include='all')
```

```python
# удалили 100% дубликаты и убедились что их там больше нет
database_films.drop_duplicates(inplace=True)
database_films.Film.value_counts()
```

```python
# проверили на наличие дубликатов у которых все поля идентичны кроме названия, увидели наличие еще одного дубликата
# у которого название написано по разному
dd = database_films.duplicated(subset=['Genre','Lead Studio','Audience score %',\
        'Profitability','Rotten Tomatoes %','Year'], keep=False)
database_films[dd]
```

```python
# сделали удаление дубликата найденного ранее
database_films.drop_duplicates(subset=['Genre','Lead Studio',\
        'Audience score %','Profitability','Rotten Tomatoes %','Year'],inplace=True)
database_films
```

```python
#  сделали преобразование Year в фотмат даты
database_films.Year = pd.to_datetime(database_films.Year, format='%Y')
database_films.head()
```

```python
# получили статистическую информацию по столбцам с типом int и float
# count - кол-во записей
# mean - среднее значение 
# std - стандартное отклонение
# min - минимальное значение в столбце
# 25%, 50%, 75% -  процентиль (% значений которые находятся ниже этого значения)
# 50% - также соответствует медиане
# max -  максимальное значение в столбце

# если указать include all, то мы получим данные и для полей типа строка
# unique - уникальных
# top - это наиболее часто встречающееся значение
# freq -это частота наиболее распространенного значения

database_films.describe()

```

```python
# вычисление ассиметрии по числовым полям
 
database_films.skew(numeric_only = True)

```

```python
# дает наиболее часто встречающиеся значения

database_films.mode(axis=0, numeric_only=True, dropna=True)
```

```python
# 6. Покажите подробную информацию о наименее и наиболее прибыльных
# фильмах в наборе данных.

database_films.loc[database_films['Profitability'] == database_films['Profitability'].max()]
```

```python
# 7. Значение «Жанр» временами кажется непоследовательным; попробуйте найти
# эти несоответствия и исправить их.

database_films.loc[:, ('Genre')] = database_films['Genre'].str.title().replace('Comdy',\
          'Comedy').replace('Romence', 'Romance')
database_films.Genre.unique()   
```

```python
# 8. Сохраните (в новый файл CSV) 10 лучших комедий по количеству зрителей;
# покажите только название фильма, год и студию
database_films.loc[database_films['Genre'] == 'Comedy'].sort_values(by=['Audience score %'],\
          ascending=False).head(10)[['Film','Lead Studio','Year']].to_csv('best_films.csv', index=False)
```
